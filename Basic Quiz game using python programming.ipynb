```python


def ask_question(question, options, correct_answer):
   "Asks a question with multiple-choice options and checks the user's answer."
   print(question)
   for i, option in enumerate(options):
       print(f"  {i + 1}. {option}")
   user_answer = input("Your answer:- ")
   while not user_answer.isdigit() or int(user_answer) not in range(1, 5):
       print("Invalid input. Please enter a number between 1 and 4.")
       user_answer = input("Your answer:- ")
   user_answer = int(user_answer) 
   if user_answer == correct_answer:
       print("Correct!")
       return True
   else:
       print("Incorrect. The correct answer is", options[correct_answer - 1])
       return False

def play_quiz():
   "Start the quiz game and keeps track of the score."
   score = 0
   questions =[
       
       {"question":"AYODHYA RAMA MANDIR is located in which state?",
        'options': ['Andhrapradesh','Uttara pradesh','Karnataka','Jharkhand'],
        'correct_answer': 2},

       {"question":"The metal whose salts are sensitive to light is?",
        'options': ['zinc','silver','cupper','Aluminum'],
        'correct_answer': 2},

        {"question":"Who among the following wrote Sanskrit grammar?",
        'options': ['panini','kalidas','Aryabhatt','charak'],
        'correct_answer': 1},

        {"question":"The country that has the highest in Barley Production?",
        'options': ['chaina','India','France','Russia'],
        'correct_answer': 4},

        {"question":"The Central Rice Research Station (CRRS) is situated in?",
        'options': ['Chennai','Quilon','Cuttack','Bangalore'],
        'correct_answer': 3},

        {"question":"Which one of the following river flows between Vindhyan and Satpura ranges?",
        'options': ['Narmada','Mahanadi','son','Netraveti'],
        'correct_answer': 1},

        
       {"question": "What is the capital of France?",
        "options": ["London", "Paris", "Berlin", "Rome"],
        "correct_answer": 2},

       {"question": "What is the highest mountain in the world?",
        "options": ["Mount Everest", "K2", "Kangchenjunga", "Lhotse"],
        "correct_answer": 1},

       {"question": "What is the largest ocean in the world?",
        "options": ["Pacific Ocean", "Atlantic Ocean", "Indian Ocean", "Arctic Ocean"],
        "correct_answer": 1}
                             ]
   for question in questions:
       if ask_question(question["question"], question["options"], question["correct_answer"]):
           score += 1

   print("Your final score is", score, "out of", len(questions))

play_quiz()







```
